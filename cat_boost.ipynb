```python
import numpy as np
import pandas as pd
```

## Importing the data set

```python
dataset = pd.read_csv('sonar_data.csv', header = None)
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

## Splitting the data set into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

## Training the CatBoost model on the Training set

```python
from catboost import CatBoostClassifier
classifier = CatBoostClassifier()
classifier.fit(X_train, y_train)
```

## Model Evaluation

### Accuracy on the Training set

```python
y_pred_train = classifier.predict(X_train)
print(np.concatenate((y_train.reshape(-1, 1), y_pred_train.reshape(-1, 1)), axis = 1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
print(confusion_matrix(y_train, y_pred_train))
print(accuracy_score(y_train, y_pred_train))
```

### Accuracy on the Test data

```python
y_pred_test = classifier.predict(X_test)
print(np.concatenate((y_test.reshape(-1, 1), y_pred_test.reshape(-1, 1)), axis = 1))
```

```python
print(confusion_matrix(y_test, y_pred_test))
print(accuracy_score(y_test, y_pred_test))
```
